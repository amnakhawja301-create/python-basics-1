#### INPUT FUNCTION : 
 in python it is used to take input from user
   whatever user types is stored in a variable 
   it always returns input as a string by default 

 #### TYPE CASTING: 
  it is used to convert one datatype to another datatype

```python
x= input("enter your age")
 
print("the age of user is ", x )

# input function  by default  returns string unless type casting is done
print ( type(x) )

#type casting will turn string into integer here

x = int( input("enter your age"))
print("the age of user is ", x )
print ( type(x) )
```
