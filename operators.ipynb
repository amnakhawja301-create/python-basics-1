 #### -> Arithematic operators
 (+, -, * % ,/ )
 

```python
# add two numbers  and print result+
num1=25
num2= 15
 
print("the sum is = ",num1+num2)

# find remiander
a= 2
b= 25
r= b % a
print("the remainder is= ", r)

# square of a number 
x= 5
square = x** 2   # **2 means raise to the power 2
print("square of x = ", square )
```

#### ->Assignment operators 
(=)


```python
#assign value to variable and increase by 5
y= 15
y+=5  # -> y=y+5
print("y is " ,y)

#multiply a variable by 2 using = operator
z=8  
z*=2  # -> z=z*2
print ("z is " ,z)

#subtract 10 from variable using = operator
num=30
num-=10
print(num)
```

#### ->Comparison operators
 (<=, >=, < , >, != ,==)


```python
#check whether two numbers are equal or not
number1= 50
number2= 4
print(number1==number2)

# check if one number is greater than other
print( number1> number2)

# compare two numbers and print whether first number
#  is less than or equal to second number
result = number2<=number1
print(result)
```

#### ->Logical operators 
(and ,or ,not)


```python
# check if a number is greater than 5 and less than 20
num= 16
print(num>5 and num<20)

# check if a number is less than 10 or greater than 50
num=55
print( num<10 or num>50)

# not operator to reverse a condition
a= 20
print ( not( a>5 and a<25)) 
```

#### ->Membership operators
 (in , not in)

```python
#check if a number exists in a given list
items= [23,24,25,28,22]
print(25 in items )

# check if a word exists in a list
words= ['A','B','C', 'D','F']
print( 'A' in words )

# check whether a value is not present in a tuple
tuple= (1,3,8,6,9,2)
print(10 not in tuple)
```
